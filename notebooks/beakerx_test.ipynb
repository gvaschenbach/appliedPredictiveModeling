```r
library( AppliedPredictiveModeling)
library( tidyverse )
```

```r
data(abalone)
```

```r
head( abalone )
```

```r
attach( abalone )
```

```r
ggplot( abalone, aes( x = WholeWeight ) ) +
    geom_density() + 
    theme_classic()
```

```r
ggplot( abalone, aes( x = WholeWeight, colour = Type ) ) +
    geom_density() + 
    theme_classic()
```

```r
fit <- glm( WholeWeight
          ~ Type
          + LongestShell
          + Height
          + Rings
          , family = gaussian() 
          , data = abalone )
```

```r
summary( fit )
```
